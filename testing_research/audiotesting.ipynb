```python
import torch
import torchaudio
import matplotlib.pyplot as plt
```

```python
filename = "sample1.mp3"
waveform, sample_rate = torchaudio.load(filename)
data = torchaudio.functional.detect_pitch_frequency(waveform=waveform, sample_rate=sample_rate)
plt.figure()
plt.plot(data.t().numpy())
```
